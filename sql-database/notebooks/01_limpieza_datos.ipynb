# Limpieza de datos — Proyecto SQL

## Proyecto
Salud y felicidad: un análisis de la relación entre los indicadores de salud y la calidad de vida

## Objetivo

Preparar los datasets para cargarlos en una base de datos SQL.

## Datasets utilizados

- Breast Cancer Wisconsin
- Heart Disease
- World Happiness Report

## Flujo de trabajo

1. Cargar datos originales
2. Explorar estructura de los datos
3. Detectar problemas de calidad
4. Limpiar columnas y valores
5. Exportar datasets limpios para SQL

```python
# ==============================
# IMPORTAR LIBRERÍAS
# ==============================

import pandas as pd
import numpy as np

# mostrar todas las columnas
pd.set_option("display.max_columns", None)

# evitar notación científica
pd.set_option("display.float_format", lambda x: f"{x:.2f}")
```

```python
# ==============================
# CARGAR DATASETS
# ==============================

breast_cancer = pd.read_csv("../data/raw/breast_cancer.csv")
heart_disease = pd.read_csv("../data/raw/heart_disease.csv")
happiness = pd.read_csv("../data/raw/world_happiness.csv")

print("Datasets cargados correctamente")
```

```python
# ==============================
# VER PRIMERAS FILAS
# ==============================

print("Breast Cancer")
display(breast_cancer.head())

print("Heart Disease")
display(heart_disease.head())

print("World Happiness")
display(happiness.head())
```

```python
# ==============================
# TAMAÑO DE LOS DATASETS
# ==============================

print("Breast Cancer shape:", breast_cancer.shape)
print("Heart Disease shape:", heart_disease.shape)
print("World Happiness shape:", happiness.shape)
```

```python
# ==============================
# COLUMNAS DE CADA DATASET
# ==============================

print("Columnas Breast Cancer:")
print(breast_cancer.columns)

print("\nColumnas Heart Disease:")
print(heart_disease.columns)

print("\nColumnas World Happiness:")
print(happiness.columns)
```

```python
# ==============================
# VALORES NULOS
# ==============================

print("Breast Cancer - valores nulos")
print(breast_cancer.isna().sum())

print("\nHeart Disease - valores nulos")
print(heart_disease.isna().sum())

print("\nWorld Happiness - valores nulos")
print(happiness.isna().sum())
```

```python
# ==============================
# LIMPIEZA DATASET BREAST CANCER
# ==============================

bc_clean = breast_cancer.copy()

# estandarizar nombres de columnas
bc_clean.columns = (
    bc_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

# eliminar columnas innecesarias si existen
bc_clean = bc_clean.loc[:, ~bc_clean.columns.str.contains("^unnamed")]

# eliminar duplicados
bc_clean = bc_clean.drop_duplicates()

# revisar resultado
bc_clean.head()
```

```python
# revisar columnas finales me daba error, así que lo hago con .tolist()
bc_clean.columns.tolist()
```

```python
# ==============================
# RECARGAR BREAST CANCER CON COLUMNAS CORRECTAS
# ==============================

columnas_breast_cancer = [
    "id",
    "diagnosis",
    "radius_mean",
    "texture_mean",
    "perimeter_mean",
    "area_mean",
    "smoothness_mean",
    "compactness_mean",
    "concavity_mean",
    "concave_points_mean",
    "symmetry_mean",
    "fractal_dimension_mean",
    "radius_se",
    "texture_se",
    "perimeter_se",
    "area_se",
    "smoothness_se",
    "compactness_se",
    "concavity_se",
    "concave_points_se",
    "symmetry_se",
    "fractal_dimension_se",
    "radius_worst",
    "texture_worst",
    "perimeter_worst",
    "area_worst",
    "smoothness_worst",
    "compactness_worst",
    "concavity_worst",
    "concave_points_worst",
    "symmetry_worst",
    "fractal_dimension_worst"
]

breast_cancer = pd.read_csv(
    "../data/raw/breast_cancer.csv",
    header=None,
    names=columnas_breast_cancer
)

breast_cancer.head()
```

```python
# ==============================
# LIMPIAR VARIABLE DIAGNOSIS
# ==============================

bc_clean["diagnosis"] = bc_clean["diagnosis"].replace({
    "M": "Maligno",
    "B": "Benigno"
})

bc_clean["diagnosis"].value_counts()
```

```python
# ==============================
# INFORMACIÓN FINAL DATASET LIMPIO
# ==============================

print("Shape:", bc_clean.shape)

print("\nValores nulos:")
print(bc_clean.isna().sum())

print("\nDuplicados:", bc_clean.duplicated().sum())
```

```python
# Crear copia limpia para empezar la limpieza
# ==============================
# COPIA PARA LIMPIEZA
# ==============================

heart_clean = heart_disease.copy()

heart_clean.head()
```

```python
# ==============================
# LIMPIAR NOMBRES DE COLUMNAS
# ==============================

heart_clean.columns = (
    heart_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

heart_clean.columns
```

```python
# ==============================
# RENOMBRAR COLUMNAS - El dataset usa nombres raros los renombraremos para hacerlos mas descriptivos y claros.
# ==============================

heart_clean = heart_clean.rename(columns={
    "trestbps": "resting_bp",
    "chol": "cholesterol",
    "fbs": "fasting_blood_sugar",
    "thalach": "max_heart_rate",
    "exang": "exercise_angina",
    "target": "heart_disease",
    "cp": "chest_pain_type"
})

heart_clean.head()
```

```python
# ==============================
# RECARGAR HEART DISEASE CON COLUMNAS CORRECTAS
# ==============================

columnas_heart_disease = [
    "age",
    "sex",
    "chest_pain_type",
    "resting_bp",
    "cholesterol",
    "fasting_blood_sugar",
    "resting_ecg",
    "max_heart_rate",
    "exercise_angina",
    "oldpeak",
    "slope",
    "ca",
    "thal",
    "heart_disease"
]

heart_disease = pd.read_csv(
    "../data/raw/heart_disease.csv",
    header=None,
    names=columnas_heart_disease,
    na_values="?"
)

heart_disease.head()
```

```python
# ==============================
# LIMPIEZA HEART DISEASE
# ==============================

heart_clean = heart_disease.copy()

heart_clean = heart_clean.drop_duplicates()

columnas_heart = [
    "age",
    "sex",
    "chest_pain_type",
    "resting_bp",
    "cholesterol",
    "fasting_blood_sugar",
    "max_heart_rate",
    "exercise_angina",
    "oldpeak",
    "heart_disease"
]

heart_clean = heart_clean[columnas_heart]

heart_clean.head()
```

```python
print("Shape:", heart_clean.shape)

print("\nValores nulos:")
print(heart_clean.isna().sum())

print("\nDuplicados:", heart_clean.duplicated().sum())
```

```python
# ==============================
# COPIA PARA LIMPIEZA - Creamos copia para limpiar el dataset de felicidad sin modificar el original.
# ==============================

happy_clean = happiness.copy()

happy_clean.head()
```

```python
# ==============================
# LIMPIAMOS NOMBRES DE COLUMNAS
# ==============================

happy_clean.columns = (
    happy_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

happy_clean.columns
```

```python
# ==============================
# ELEGIMOS NUESTRAS VARIABLES
# ==============================

columnas_happy = [
    "country_or_region",
    "score",
    "gdp_per_capita",
    "social_support",
    "healthy_life_expectancy"
]

happy_clean = happy_clean[columnas_happy]

happy_clean.head()
```

```python
# ==============================
# RENOMBRAR COLUMNAS - esto lo hacemos para que SQL sea mas claro. 
# ==============================

happy_clean = happy_clean.rename(columns={
    "country_or_region": "country",
    "score": "happiness_score"
})

happy_clean.head()
```

```python
print("Shape:", happy_clean.shape)

print("\nValores nulos:")
print(happy_clean.isna().sum())

print("\nDuplicados:", happy_clean.duplicated().sum())
```
